```python
import cv2
import os
import numpy as np
from scipy import ndimage
import SimpleITK as sitk 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 读取图像并转换为灰度图
imageori_color = cv2.imread('img_origin.png')
imageori_gray = cv2.cvtColor(imageori_color, cv2.COLOR_BGR2GRAY)


image_gray = imageori_gray[355:365,447:457]
plt.imshow(image_gray, cmap='gray')


sns.set_theme()
# Draw a heatmap with the numeric values in each cell
plt.subplots(figsize=(9, 6))
sns.heatmap(image_gray, annot=True, fmt="d", linewidths=.5,cmap="coolwarm")
plt.show()

plt.imshow(imageori_gray, cmap='gray')
plt.plot(447,355, 'ro')
plt.show()
```
