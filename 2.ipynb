```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

df1=pd.read_excel('副本监测数据.xlsx',sheet_name=0)
df2=pd.read_excel('副本监测数据.xlsx',sheet_name=1)
df3=pd.read_excel('副本监测数据.xlsx',sheet_name=2)
df4=pd.read_excel('副本监测数据.xlsx',sheet_name=3)
df1 =df1.to_numpy()
df2 =df2.to_numpy()
df3 =df3.to_numpy()
df4 =df4.to_numpy()
df = [df1,df2,df3,df4]
df
```

```python

smelist1=[]
srmelist1=[]
maelist1=[]
mapelist1=[]
smapelist1=[]
r_squaredlist1=[]
for dataframe in df:
    smelist=[]
    srmelist=[]
    maelist=[]
    mapelist=[]
    smapelist=[]
    r_squaredlist=[]
    for k in range(1,dataframe.shape[1]):
        sme = mean_squared_error(dataframe[:, 0], dataframe[:, k])
        srme = np.sqrt(sme)
        mae = mean_absolute_error(dataframe[:, 0], dataframe[:, k])
        mape = np.mean(np.abs((dataframe[:, 0] - dataframe[:, k]) / dataframe[:, 0]))
        smape = np.mean(np.abs(dataframe[:, 0] - dataframe[:, k]) / np.abs((dataframe[:, 0] + dataframe[:, k]) / 2))
        r_squared = r2_score(dataframe[:, 0], dataframe[:, k])

        smelist.append(sme)
        srmelist.append(srme)
        maelist.append(mae)
        mapelist.append(mape)
        smapelist.append(smape)
        r_squaredlist.append(r_squared)

    smelist1.append(smelist)
    srmelist1.append(srmelist)
    maelist1.append(maelist)
    mapelist1.append(mapelist)
    smapelist1.append(smapelist)
    r_squaredlist1.append(r_squaredlist)

# 将列表转换为NumPy数组
smelist1 = np.array(smelist1)
srmelist1 = np.array(srmelist1)
maelist1 = np.array(maelist1)
mapelist1 = np.array(mapelist1)
smapelist1 = np.array(smapelist1)
r_squaredlist1 = np.array(r_squaredlist1)

# 打印结果
print("SME List:\n", smelist1)
print("SRMSE List:\n", srmelist1)
print("MAE List:\n", maelist1)
print("MAPE List:\n", mapelist1)
print("SMAPE List:\n", smapelist1)
print("R-squared List:\n", r_squaredlist1)
```

```python
results_df = pd.DataFrame({
    'SME': smelist1.flatten(),
    'SRMSE': srmelist1.flatten(),
    'MAE': maelist1.flatten(),
    'MAPE': mapelist1.flatten(),
    'SMAPE': smapelist1.flatten(),
    'R-squared': r_squaredlist1.flatten()
})

# 保存到CSV文件
results_df.to_csv("results.csv", index=False)
```
